## Payfield $V$

```python
# Make symbolic payfield from symbolic payfunctions by taking individual differentials

def make_payfield(payfuncs):
    
    return sp.Matrix([sp.diff( payfuncs[i], xi ) for i, xi in enumerate(simvars)])
```

```python
V = make_payfield(u)
V
```

## Payoff 2-form $dV$

```python
# 2-form, that is dV where V is payoff 1-form

def make_dV(payfield):
    J = payfield.jacobian(simvars)
    # A = 1/2*(J - J.T)
    # dV = 2 * A
    dV = J.T - J
    return dV
```

```python
dV = make_dV( make_payfield(u) )
dV
```

```python
# Lambdify payfield; payfield is sympy matrix

def lambdify_vector(payfield):
    return sp.lambdify(simvars, list(payfield) )
```

```python
lambdify_vector( V ) (1, 1, 1)
```

```python
# sympy matrix
def make_SGA_correction(payfield):
    V = payfield # Sympy vector
    dV = make_dV(payfield) # Simpy matrix
    return dV * V
```

```python
dVV = make_SGA_correction(V)
dVV
```

```python
lambdify_vector(dVV)(1,1,1)
```

## Spherical games

- Incompressible with respect to entropic regularizer, killing separately 2 pieces of codifferential
- $\partial_i V_i = 0$
- parallel to sphere centered at $1/2$

```python
# sympy matrix
def make_spherical_payfield_sym(spherical_payfield_1, spherical_payfield_2):
    return (spherical_payfield_1 * sp.Matrix([0,  z -1/2, - (y-1/2) ]) + spherical_payfield_2 *  sp.Matrix([ (z-1/2), 0, - (x-1/2) ]))
```

```python
# Make entropic-incompressible by parallel to radial field with center 1/2

#---------------------------------
# ANY function of x only
#---------------------------------
spherical_payfield_1 = -1
#spherical_payfield_1 = x
#spherical_payfield_1 = sp.cos(x**2) # x**2

#---------------------------------
# ANY function of x only
#---------------------------------
spherical_payfield_2 = 1
#spherical_payfield_2 = y
#spherical_payfield_2 =  sp.sin(y + y**2) # sp.exp(y) * sp.sin(y)
```

```python
spherical_payfield_sym = make_spherical_payfield_sym(spherical_payfield_1, spherical_payfield_2)
spherical_payfield = lambdify_vector( spherical_payfield_sym )
```

## Symplectic Gradient Adjustment - Euclidean, unconstrained

```python
# Only euclidean at the moment (so precisely SGA of letcher and balduzzi); to include sharp to make into symplectic FTRL
def SGA(x, time, payfield, sga_correction, lam):
    return np.array(payfield(*x)) + lam * np.array(sga_correction(*x))
```

```python
SGA( [0.5, 0.4, 0.3], 0, spherical_payfield, spherical_sga_correction, +0.5  )
```

```python
def dummy(ax):
    pass
```

```python
def make_plots_SGA(time, payfield, sga_correction, lam, initial_points, name, plot_point = False, special_point = 0, hook_plot = dummy ):
    
    sols = [ odeint(SGA, p, time, args=(payfield, sga_correction, lam)) for p in initial_points ]

    # ----------------------------------------------------
    fig = plt.figure(figsize = (8,8) )
    ax = fig.add_subplot(111, projection='3d')
    
    [ax.plot(sol[:, 0], sol[:, 1], sol[:, 2], 'k', linewidth=0.5) for sol in sols]
    [ax.plot(sol[:, 0][-1], sol[:, 1][-1], sol[:, 2][-1], 'ro', linewidth=0.5) for sol in sols]
   
    
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z", labelpad=-10)  # Increase padding for the z-label
    plt.tight_layout()

    #ax.set_xlim([0, 1])
    #ax.set_ylim([0, 1])
    #ax.set_zlim([0, 1])
    ax.set_title(f"{name}")

    [ax.plot(*p, 'ko') for p in initial_points ]

    if plot_point:
        ax.plot(*special_point, "ko")
    plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)

    hook_plot(ax)
    
    plt.show()
        
    
```

## Critical points: $x = y, z = 0.5$

```python
def hook(ax):
    # Define the range of y values
    y_crit = np.linspace(0, 1, 100)  # Adjust range and resolution as needed

    # Define x and z based on the solution
    x_crit = y_crit  # From the solution {x: y}
    z_crit = np.full_like(y_crit, 0.5)  # z is constant at 0.5
    ax.plot(x_crit, y_crit, z_crit)
```
