## Kernels

```python
#---------------------------------------------------------------------
# Define (symbolic) kernel functions andtheir labels
#---------------------------------------------------------------------

# 0 = sha
# 1 = log barrier
# 2 = q-replicator
# 3 = q-replicator

# q-replicator parameters
Q = [] # [0.5, 1.5] #, 2.5]



# Kiwiel kernel parameter; very similar to q-replicator
kiw = 0.5

# Teboulle kernel --> very similar to q-replicator

# sympy expressions
simkernels = [
    x * sp.log(x),
    - sp.log(x),
    (x**2 - x)/4,
    - (x**kiw) / kiw] + [(x - x**q) / ( 1-q ) for q in Q ]

# kernel labels
kerlabels = ["Sha", "log", "Eu", "Kiw"] + [f"q = {q} Sha" for q in Q]
#---------------------------------------------------------------------

# simkernels = simkernel + [  - sp.log(x) ]
# reglabels = reglabels  + [ "log" ]
```

## Continuous games

```python
# Example: Any payoff function for a [0,1] x [0,1] x [0,1] game

u = [x*y, -x*y, 0]
u
```

```python
# Make symbolic payfield from symbolic payfunctions by taking individual differentials

def make_payfield(payfuncs):
    
    return [sp.diff( payfuncs[i], xi ) for i, xi in enumerate(simvars)]
```

```python
# Lambdify payfield

def lambdify_payfield(payfield):
    return sp.lambdify(simvars, payfield )
```

### Spherical game

incompressible with respect to entropic regularizer, killing separately 2 pieces of codifferential --> $\partial_i V_i = 0$ and parallel to sphere centered at $1/2$

```python
def make_spherical_payfield_sym(spherical_payfield_1, spherical_payfield_2):
    return (spherical_payfield_1 * sp.Matrix([0,  z -1/2, - (y-1/2) ]) + spherical_payfield_2 *  sp.Matrix([ (z-1/2), 0, - (x-1/2) ]))
```

```python
# Make entropic-incompressible by parallel to radial field with center 1/2

# ANY function of x only
#spherical_payfield_1 = -1
spherical_payfield_1 = x
#spherical_payfield_1 = sp.cos(x**2) # x**2

# ANY function of y only
#spherical_payfield_2 = 1
spherical_payfield_2 = y
#spherical_payfield_2 =  sp.sin(y + y**2) # sp.exp(y) * sp.sin(y)



spherical_payfield_sym = make_spherical_payfield_sym(spherical_payfield_1, spherical_payfield_2)
spherical_payfield = lambdify_payfield( [v for v in spherical_payfield_sym] )
```

## Dynamics

## Initial points

```python
epsilon = 1e-3

init_generator = np.linspace(epsilon, 1-epsilon, 2)

initial_points = [ p for p in itertools.product(init_generator, init_generator, init_generator) ]
initial_points
```

```python
def ftrl(var, time, sharp, payfield):
        
    x1, x2, x3 = var
    V1, V2, V3 = payfield(x1, x2, x3)
        
    F1 = sharp(x1) * V1
    F2 = sharp(x2) * V2
    F3 = sharp(x3) * V3
        
    return [F1, F2, F3]
```

## Experiments

```python
# defalut ftrl
def make_plots(sharp_index, payfield, initial_points, time, name, plot_point = False, special_point = 0):
    
    sols = [ odeint(ftrl, p, time, args=(A.sharps[sharp_index], payfield)) for p in initial_points ]

    fig = plt.figure(figsize = (8,8) )
    ax = fig.add_subplot(111, projection='3d')
    
    [ax.plot(sol[:, 0], sol[:, 1], sol[:, 2], 'k', linewidth=0.5) for sol in sols]
   
    
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z", labelpad=-10)  # Increase padding for the z-label
    plt.tight_layout()

    ax.set_xlim([0, 1])
    ax.set_ylim([0, 1])
    ax.set_zlim([0, 1])
    ax.set_title(f"{name} - {A.reglabels[sharp_index]}")

    [ax.plot(*p, 'ko') for p in initial_points ]

    if plot_point:
        ax.plot(*special_point, "ko")
    plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)

    plt.show()
        
    
```

```python
# switch between coords and points
def coords_points(L):

	#https://stackoverflow.com/questions/2921847/what-does-the-star-operator-mean-in-a-function-call
	def make(*values):
		return values

	return make(*zip(*L))
```

```python
time = np.linspace(0, 100, 100000)


initial_points = [[0.91, 0.81, 0.31], [0.2, 0.4, 0.6]]

# harmonic, sha
make_plots(sharp_index = 0, payfield = lambdify_payfield(H.clean_payfield), initial_points = initial_points , time = time,  name = H.name)
```

```python
time = np.linspace(0, 100, 100000)
# harmonic, sha
make_plots(sharp_index = 1, payfield = lambdify_payfield(H.clean_payfield), initial_points = [[0.9, 0.8, 0.3]], time = time,  name = H.name)
```
