# Spherical Games & Symplectic FTRL

## Payoff 2-form dV

```python
# 2-form, that is dV where V is payoff 1-form

def make_pay2field(payfield):
    J = sp.Matrix(payfield).jacobian(simvars)
    # A = 1/2*(J - J.T)
    # dV = 2 * A
    dV = J.T - J
    return dV
```

```python
# Lambdify payfield

def lambdify_payfield(payfield):
    return sp.lambdify(simvars, payfield )
```

# Here magic happens: Symplectic correction to FTRL
## Symplectic Gradient Adjustment - Euclidean, unconstrained

Still only Euclidean, that is precisely SGA of Letcher and Balduzzi: run Euclidean FTRL (unconstrained) on $$V \leftarrow V + \lambda dV \cdot V$$, i.e.

$$ \dot{x} = V + \lambda dV \cdot V$$

To do nect: Symplectic FTRL: run FTRL (non-Euclidean & constrained) on  

$$\langle V(x), x-q \rangle = 0$$ 


with
- $dV$ 2-form
- $V^{\sharp}$ vector field
- $dV(V^{\sharp})$ 1-form
- $\sharp = g^-1$ sharp wrt usual primal metric
- $g = \text{hess}h$ usual primal metric
- $\implies \sharp = \text{hess}h^{\ast}$ hence sharp wrt primal metric is flat wrt dual metric

```python
# Lambdify payfield

def lambdify_SGA_correction(payfield):
    V = sp.Matrix(payfield) # Sympy vector
    dV = make_pay2field(payfield) # Simpy matrix
    dVV_sym = dV * V
    print(dVV_sym)
    dVV = list(dVV_sym) # List
    
    return sp.lambdify(simvars, dVV )
```

```python
def ftrl(var, time, sharp, payfield):
        
    x1, x2, x3 = var
    V1, V2, V3 = payfield(x1, x2, x3)
        
    F1 = sharp(x1) * V1
    F2 = sharp(x2) * V2
    F3 = sharp(x3) * V3
        
    return [F1, F2, F3]
```


## Fenchel gradient descent --> deprecated

My idea to look at dynamics that knows nothing og game and by construction converges to set of points in duals space such that $Q(y) = q$: gradient descent on Fenchel coupling

$d_y F_q = Q(y) - q$; dynamics: $\dot{y} = q - Q(y)$


```python
def logitChoice(y):
    #return np.exp(y) / np.sum(np.exp(y))  # constrained on simplex
    return np.exp(y) / (np.exp(y) + 1)     # unconstrained
```

```python
logitChoice([0.9, 0.8, 0.3])
```

```python
# Just for ENTROPIC regularizer; choice is logit, constrained on simplex or unconstrained

# q_fenchel = [0.2, 0.3, 0.5]

# if simplex choice, sum must be one; else anything on [0,1]^3
# assert sum(q_fenchel) == 1

# unconstrained
q_fenchel = [0.5, 0.5, 0.5]

def fenchelGradient(y, time):
        
   return np.array( q_fenchel ) - logitChoice(y)
```

## Random perpendicular --> deprecated

### Next: attempt, doomed to fail, to correct FTRL by approximation of negative gradient of Fenchel coupling that does not require knowledge of $q$

Need something perpendiculat to level set of Fenchel without knowing Fenchel; Knowing $V \circ Q$ can take somethibg perpendicular to that, but it's clearly not enough - trying to find a vector perpendicular to a surface given only a vector parallel to the surface. Got 1 constraints; need N-1 constraints. Dimensionally impossible, do not pursue this way. Rather --> **Symplectic FTRL correction**

```python
def find_perpendicular(vector):
    n = len(vector)
    # Create a random vector of the same dimension
    random_vector = np.random.rand(n)
    # Solve for a perpendicular vector
    perpendicular = random_vector - np.dot(random_vector, vector) / np.dot(vector, vector) * np.array(vector)
    return np.array(perpendicular)

# Example usage:
v = np.array([1, 2, 3])
perp_v = find_perpendicular(v)
print(f"Original vector: {v}, Perpendicular vector: {perp_v}")
```

```python
import numpy as np

def perpendicular_3d(vector):
    # Choose a different vector to take the cross product with.
    # This example uses [1, 0, 0], but ensure it's not parallel to the input vector.
    x, y, z = vector
    if x == 0 and y == 0:  # If the input vector is along the z-axis
        other = [1, 0, 0]
    else:
        other = [0, 0, 1]  # Default to z-axis
    return np.cross(vector, other)

# Example usage:
v = np.array([1, 4, 3])
perp_v = perpendicular_3d(v)
print(f"Original vector: {v}, Perpendicular vector: {perp_v}")
```

```python
def perturbed_payfield(payfield, x):
    Vx = payfield(*x)
    eps = 1
    return Vx * np.random.uniform(low=-eps, high=eps, size=np.array(Vx).shape)
```

```python
def fenchelFTRLperp(y, time, payfield, FTRL_coeff, Fenchel_coeff):
    
    x = logitChoice(y)
    
    VQ = np.array(payfield( *x ))
    VQp = np.array(perturbed_payfield( payfield, x ))
    

    
    return FTRL_coeff * VQ + Fenchel_coeff * np.cross(VQ, VQp)
```

```python
# switch between coords and points
def coords_points(L):

	#https://stackoverflow.com/questions/2921847/what-does-the-star-operator-mean-in-a-function-call
	def make(*values):
		return values

	return make(*zip(*L))
```

```python
def make_fenchel_plots(initial_points, time, name, plot_point = False, special_point = 0, special_label = 0):
    
    sols = [ odeint(fenchelGradient, p, time) for p in initial_points ]

    fig = plt.figure(figsize = (12,16) )
    dual_ax = fig.add_subplot(121, projection='3d')

    Y1 = [sol[:, 0] for sol in sols]
    Y2 = [sol[:, 1] for sol in sols]
    Y3 = [sol[:, 2] for sol in sols]

    DUAL_COORDS = [  [ Y1[p], Y2[p], Y3[p] ] for p in range(len(sols))]
    DUAL_POINTS = [ coords_points(coords) for coords in DUAL_COORDS ]

    
    [dual_ax.plot(*coords, 'k', linewidth=0.5) for coords in DUAL_COORDS]
    dual_ax.set_xlabel("y1")
    dual_ax.set_ylabel("y2")
    dual_ax.set_zlabel("y3", labelpad=-10)  # Increase padding for the z-label
    dual_ax.set_title(f"{name} - Dual")

    # ---------------------------------------------------------------
    primal_ax = fig.add_subplot(122, projection='3d')
    PRIMAL_POINTS = [[logitChoice(y) for y in dual_points] for dual_points in DUAL_POINTS]
    PRIMAL_COORDS = [coords_points(points) for points in PRIMAL_POINTS]
    [primal_ax.plot(*coords, 'k', linewidth=0.5) for coords in PRIMAL_COORDS]

    primal_ax.set_xlim([0, 1])
    primal_ax.set_ylim([0, 1])
    primal_ax.set_zlim([0, 1])
    
    [primal_ax.plot(*logitChoice(p), 'ko') for p in initial_points ]
    primal_ax.set_xlabel("x1")
    primal_ax.set_ylabel("x2")
    primal_ax.set_zlabel("x3", labelpad=-10)  # Increase padding for the z-label
    primal_ax.set_title(f"{name} - Primal")

    if plot_point:
        primal_ax.plot(*special_point, "ro", label = special_label)
    plt.legend(loc = 'best')
    #plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)

    plt.tight_layout()
    plt.show()
        
    
```

```python
# codifferential of payoff 1-form
def codiff(sim_payfield, regularizer_index):

    h2 = A.simregularizers2[regularizer_index]
    h3 = A.simregularizers3[regularizer_index]
    
    sum_1 = sp.simplify(sum([   sp.diff( sim_payfield[i], simvars[i] ) / h2.subs(x, simvars[i]) for i in [0,1, 2]   ]))
    sum_2 = sp.simplify(sum( [   - 1 / 2 * sim_payfield[i] * (sp.simplify(h3 / (h2**2)).subs(x, simvars[i])) for i in [0,1, 2]   ] ))
    
    print("Two individual pieces:" , sum_1, sum_2)
    return sum_1 + sum_2
```

```python
# codifferential of payoff 1-form, allowing different regularizers for different players
def codiff_different_regularizers(sim_payfield, regularizer_indices):

    assert len(sim_payfield) == len(regularizer_indices)

    h2 = [A.simregularizers2[i] for i in regularizer_indices]
    h3 = [A.simregularizers3[i] for i in regularizer_indices]
    
    sum_1 = sp.simplify(sum([   sp.diff( sim_payfield[i], simvars[i] ) / h2[i].subs(x, simvars[i]) for i in [0,1, 2]   ]))
    sum_2 = sp.simplify(sum( [   - 1 / 2 * sim_payfield[i] * (sp.simplify(h3[i] / (h2[i]**2)).subs(x, simvars[i])) for i in [0,1, 2]   ] ))
    
    print("Two individual pieces:" , sum_1, sum_2)
    return sum_1 + sum_2
```
