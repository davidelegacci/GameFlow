## Kernels

```python
#---------------------------------------------------------------------
# Define (symbolic) kernel functions andtheir labels
#---------------------------------------------------------------------

# 0 = sha
# 1 = log barrier
# 2 = q-replicator
# 3 = q-replicator

# q-replicator parameters
Q = [] # [0.5, 1.5] #, 2.5]



# Kiwiel kernel parameter; very similar to q-replicator
kiw = 0.5

# Teboulle kernel --> very similar to q-replicator

# sympy expressions
simkernels = [
    x * sp.log(x),
    (x**2 - x)/4,
    - (x**kiw) / kiw] + [(x - x**q) / ( 1-q ) for q in Q ]

# kernel labels
kerlabels = ["Sha", "Eu", "Kiw"] + [f"q = {q} Sha" for q in Q]
#---------------------------------------------------------------------

# simkernels = simkernel + [  - sp.log(x) ]
# reglabels = reglabels  + [ "log" ]
```

## Continuous games

```python
# Example: Any payoff function for a [0,1]^N game

w1, w2, w3, w4, w5 = simvars
```

```python
# Make symbolic payfield from symbolic payfunctions by taking individual differentials

def make_payfield(payfuncs):
    
    return [sp.diff( payfuncs[i], xi ) for i, xi in enumerate(simvars)]
```

```python
# Lambdify payfield

def lambdify_payfield(payfield):
    return sp.lambdify(simvars, payfield )
```

## Spherical game

## incompressible with respect to entropic regularizer, killing separately 2 pieces of codifferential --> $\partial_i V_i = 0$ and parallel to sphere centered at $1/2$

```python
simvars
```

```python
def make_spherical_payoff_sym(functions):
    
    assert len(functions) == N-1
    pays = [ ]
    last = [ ]
    q = 0.5
    
    for i, f in enumerate(functions):
        assert sp.diff(f, simvars[i]) == 0, f"{i+1} depends on itself"
        assert sp.diff(f, simvars[-1]) == 0, f"{i+1} depends on last"
        pays.append( f * (simvars[-1] - q) * simvars[i] )
        last.append( f * (simvars[i] - q) )

    pays.append(   - sum(last) * simvars[-1]    )
    return([sp.simplify(u) for u in pays])
```

```python
def changer(factor, a,b):
    return -factor * a*b * (0.5-b*b) * (1-b*b)
```

## Payoff

```python
# HERE payoff DOFS

functions = [w2*w3+w4, changer(10, w1, w3), changer(5, w2, w1), w1**2*(1-w2*w3*w1) ]

spherical_payoff_sym = make_spherical_payoff_sym(functions)
```

```python
spherical_payoff_sym
```

```python
spherical_payfield_sym = [sp.simplify(v) for v in make_payfield(spherical_payoff_sym)]
```

```python
def ftrl(var, time, sharp, payfield):
        
    #x1, x2, x3 = var
    #V1, V2, V3 = payfield(x1, x2, x3)
        
    #F1 = sharp(x1) * V1
    #F2 = sharp(x2) * V2
    #F3 = sharp(x3) * V3
        
    return [ sharp( var[i] ) * payfield(*var)[i] for i in range(len(var)) ]
```

## Experiments

```python
def make_plots(sharp_index, payfield, initial_points, time, name, plot_point = False, special_point = 0, dims_to_plot = [0,1,2]):
    
    sols = [ odeint(ftrl, p, time, args=(A.sharps[sharp_index], payfield)) for p in initial_points ]

    fig = plt.figure(figsize = (8,8) )
    ax = fig.add_subplot(111, projection='3d')

    dim_x, dim_y, dim_z = dims_to_plot
    
    [ax.plot(sol[:, dim_x], sol[:, dim_y], sol[:, dim_z], 'k', linewidth=0.5) for sol in sols]
   
    
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z", labelpad=-10)  # Increase padding for the z-label
    plt.tight_layout()

    ax.set_xlim([0, 1])
    ax.set_ylim([0, 1])
    ax.set_zlim([0, 1])
    ax.set_title(f"{name} - {A.reglabels[sharp_index]} - Plotting {dims_to_plot}")

    [ax.plot(*[p[i] for i in dims_to_plot], 'ko') for p in initial_points ]

    if plot_point:
        ax.plot(*special_point, "ko")
    plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)

    plt.show()
        
    
```

```python
time = np.linspace(1, 1000, 10000)
initial_points = [[0.2, 0.3, 0.2, 0.4, 0.6]]
for triple in triples:
    make_plots(sharp_index = 0, payfield = spherical_payfield, initial_points = initial_points, time = time, name = "Spherical", dims_to_plot = triple)
```
