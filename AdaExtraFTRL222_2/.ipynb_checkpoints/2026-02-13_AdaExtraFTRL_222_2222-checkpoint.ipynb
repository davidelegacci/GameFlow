##  AdaExtraFTRL

```python

# effective mirror
def mirrori(y):
    return np.exp(y) / (1 + np.exp(y)  )

#mirror = [mirrori for _ in range(EFFECTIVE_DIMENSION)]
```

```python
def make_signal(coef, v_curr, v_beforelead):
    return coef * v_curr + (1 - coef) * v_beforelead
```

```python
# Domain = [0,1]
# Quadratic losses

def adaFTRLplus(y_1, nRuns, eta_1 = np.ones(EFFECTIVE_DIMENSION), v_half = np.zeros(EFFECTIVE_DIMENSION), coef = 0.5 * np.ones(EFFECTIVE_DIMENSION) ):

    "y_1 = np.array([a, b, c, d])"
    
    x_1 = mirrori(eta_1 * y_1)
    v_1 = payfield(*x_1)
    signal_1 = make_signal(coef, v_1, v_half) 
    t = 1

    y_curr = y_1
    eta_curr = eta_1

    x_curr = x_1
    v_curr = v_1
    signal_curr = signal_1

    y_lead = np.zeros(EFFECTIVE_DIMENSION)
    x_lead = np.zeros(EFFECTIVE_DIMENSION)
    v_lead = np.zeros(EFFECTIVE_DIMENSION)
    g_sum_curr = np.zeros(EFFECTIVE_DIMENSION)
    

    
    lead_sequence = []
    
    #utility_sequence = []
    
    while t <= nRuns:

        y_lead = y_curr + signal_curr
        x_lead = mirrori( eta_curr * y_lead )
        lead_sequence.append(x_lead)
        
        v_lead = payfield(*x_lead)
        

        g_curr = v_lead - signal_curr
        
        #g_sum_curr += (npla.norm( g_curr, ord = np.inf ))**2 # no norm needed, g is 1-dimensional for each player
        g_sum_curr += g_curr * g_curr
        
        eta_next = (1 + g_sum_curr)**(-1/2)
        y_next = y_curr + v_lead
        x_next = mirrori( eta_next * y_next )
        v_next = payfield(*x_next)
        signal_next = make_signal(coef, v_next, v_lead)
    
        t+=1

        y_curr = y_next
        eta_curr = eta_next
        signal_curr = signal_next
        
        
    return np.array(lead_sequence)
```

```python
# y_1 =  np.array([1, 2, 3, 4])
# y_1 = 50 * np.random.rand(EFFECTIVE_DIMENSION)

NUM_INITIAL_POINTS = 100

y_1s = [np.random.randint(-3, 3, EFFECTIVE_DIMENSION ) for _ in range(NUM_INITIAL_POINTS)]

points_ada = [adaFTRLplus(y_1, nRuns = 5000, eta_1 = np.ones(EFFECTIVE_DIMENSION), v_half = np.zeros(EFFECTIVE_DIMENSION), coef = 0.5 * np.ones(EFFECTIVE_DIMENSION) ) for y_1 in y_1s]
```

### Effective replicator dynamics for [0,1]^N game

$$ \dot{x}_i = \frac{V_i(x)}{h''_i(x_i)} = x_i (1-x_i) V_i(x)$$

```python
def da(var, time):
        
    #x1, x2, x3 = var
    #V1, V2, V3 = payfield(x1, x2, x3)
        
    #F1 = sharp(x1) * V1
    #F2 = sharp(x2) * V2
    #F3 = sharp(x3) * V3
        
    return [ var[i] * (1 - var[i]) * payfield(*var)[i] for i in range(len(var)) ]
```

### Solve ODE

```python


from scipy.integrate import odeint

time = np.linspace(1, 100, 2000)

NUM_INITIAL_POINTS = 1


#initial_points = [ (0.55, 0.45, 0.01), [0.45, 0.2, 0.55], [0.001, 0.001, 0.001]  ]  + [np.random.rand(3) for _ in range(20)]
#initial_points = [np.random.rand(EFFECTIVE_DIMENSION) for _ in range(NUM_INITIAL_POINTS)]
initial_points = [ np.array(BARYCENTER) + np.array([0.4, 0.2, 0.15, 0.45]) ]
#initial_points = [ (0.5, 0.1, 0.5)]

sols_ode = [ odeint(da, p, time) for p in initial_points ]

# sols is list of POINTS for each initial point, that is
# len(sols) = # initial points
# len( sols[0] ) = # time intervals (time series of points)
# len (sols[0][0] ) = effective dimension
```
