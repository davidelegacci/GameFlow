## Normalized harmonic measures, full representation

All the measures making the game harmonic, normalized. As it turns out, the solution space has dimension 2.

```python
## KEY: normalized measures manifold in parametric form: (x, y, z) = func( mu21, mu22 ), with x, y, z in full representatin (sum up to 1)

mu = [sp.Matrix([
[                 mu21/(1.2*mu21 + 0.8*mu22)],
[(0.2*mu21 + 0.8*mu22)/(1.2*mu21 + 0.8*mu22)]]), sp.Matrix([
[(0.8*mu21 + 0.2*mu22)/(3.2*mu21 - 1.2*mu22)],
[(2.4*mu21 - 1.4*mu22)/(3.2*mu21 - 1.2*mu22)]]), sp.Matrix([
[mu21/(mu21 + mu22)],
[mu22/(mu21 + mu22)]])]
```

```python
# x(mu21, mu22) full

mu[0]
```

```python
# y(mu21, mu22) full

mu[1]
```

```python
# z(mu21, mu22) full

mu[2]
```

## Normalized harmonic measures, effective representation

Get effective $(x, y, z) = f(\mu_{21}, \mu_{22})$ and plot parametric surface

```python
# Effective representation: (x, y, z) = func(mu21, mu22), keeping one coordinate only
index_to_keep = 1

mu_eff = [  mui[index_to_keep] for mui in mu  ]
mu_eff
```
