## Measures

Use `find_harmonic_measure` script to find **all** the measures making the game harmonic. As it turns out, the measure space has dimension 2 - it's a surface, not a curve.

Copy-paste here the solution.

```python
def make_symbols(N,s, shift = 0):
	my_symbols = []
	for i in range(shift, N + shift):
	    tmp_st = f'{s}{i}'
	    globals()[tmp_st] = sp.Symbol(tmp_st)
	    my_symbols.append(globals()[tmp_st])
	return my_symbols

mu_symbols = [ make_symbols(skeleton[i], f'mu{i}', shift = 1) for i in game.players ]
```

```python
mu_symbols
```

```python
## KEY: measures manifold in parametric form: (x, y, z) = func( mu21, mu22 ), with x, y, z in full representatin (sum up to 1)

mu = [sp.Matrix([
[                 mu21/(1.2*mu21 + 0.8*mu22)],
[(0.2*mu21 + 0.8*mu22)/(1.2*mu21 + 0.8*mu22)]]), sp.Matrix([
[(0.8*mu21 + 0.2*mu22)/(3.2*mu21 - 1.2*mu22)],
[(2.4*mu21 - 1.4*mu22)/(3.2*mu21 - 1.2*mu22)]]), sp.Matrix([
[mu21/(mu21 + mu22)],
[mu22/(mu21 + mu22)]])]
```

```python
# x(mu21, mu22) full

mu[0]
```

```python
# y(mu21, mu22) full

mu[1]
```

```python
# z(mu21, mu22) full

mu[2]
```

```python
# Effectove representation: (x, y, z) = func(mu21, mu22), keeping one coordinate only
index_to_keep = 1

mu_eff = [  mui[index_to_keep] for mui in mu  ]
mu_eff
```

```python
# x(mu21, mu22) effective
mu_eff[0]
```

```python
# y(mu21, mu22) effective
mu_eff[1]
```

```python
# z(mu21, mu22) effective
mu_eff[2]
```

```python
# Lambdify function (x, y, z) = func( mu21, mu22 ) and plot parametric surface clipped to cube
mu_func = sp.lambdify( [mu21, mu22], mu_eff )
```

```python
# Check that uniform measure (each mu = 1) yields uniformly mixed strategy (0.5, 0.5, 0.5)
mu_func(1, 1)
```

```python
# Set range to plot parametric surface; whatever, will clip to cube.
mu_range = np.linspace(0.01, 10, 1000)
```

```python
# "Mu-Nash" points: (x, y, z) on parametric surface given by feasible measures --> numerical issues!
MU1, MU2 = np.meshgrid(mu_range, mu_range)
Mu_Nash_x_unclipped, Mu_Nash_y_unclipped, Mu_Nash_z_unclipped = mu_func( MU1, MU2 )
```

```python
# Clip to cube
mu_mask = (
    (Mu_Nash_x_unclipped >= 0) & (Mu_Nash_x_unclipped <= 1) &
    (Mu_Nash_y_unclipped >= 0) & (Mu_Nash_y_unclipped <= 1) &
    (Mu_Nash_z_unclipped >= 0) & (Mu_Nash_z_unclipped <= 1)
)
```

```python
# Clip to cube
Mu_Nash_x_surf = np.where(mu_mask, Mu_Nash_x_unclipped, np.nan)
Mu_Nash_y_surf = np.where(mu_mask, Mu_Nash_y_unclipped, np.nan)
Mu_Nash_z_surf = np.where(mu_mask, Mu_Nash_z_unclipped, np.nan)
# These values are in mesh format, suitable for parametric surface plot. Used below in ax.plot_surface(Mu_Nash_x_surf, Mu_Nash_y_surf, Mu_Nash_z_surf)
```

## Solve ODE, and plot trajectories

### Effective Dual averaging with entropic regularizer for [0,1]^N game

$$ \dot{x}_i = \frac{V_i(x)}{h''_i(x_i)} = x_i (1-x_i) V_i(x)$$

```python
def da(var, time):
        
    #x1, x2, x3 = var
    #V1, V2, V3 = payfield(x1, x2, x3)
        
    #F1 = sharp(x1) * V1
    #F2 = sharp(x2) * V2
    #F3 = sharp(x3) * V3
        
    return [ var[i] * (1 - var[i]) * payfield(*var)[i] for i in range(len(var)) ]
```

### Solve ODE

```python

from scipy.integrate import odeint

time = np.linspace(1, 100, 10000)



initial_points = [ (0.55, 0.45, 0.01), [0.45, 0.2, 0.55], [0.001, 0.001, 0.001]  ]  + [np.random.rand(3) for _ in range(20)]
#initial_points = [np.random.rand(3) for _ in range(30)]
#initial_points = [ (0.51, 0.51, 0.01)]
#initial_points = [ (0.5, 0.1, 0.5)]

sols = [ odeint(da, p, time) for p in initial_points ]
```

### Plot; response graph comes from nfg module; dynamics comes from here

```python
name = 'harmonic-nash-manifold'

fig = plt.figure(figsize=(18*1.3, 12*1.3))
ax = fig.add_subplot(1, 1, 1, projection='3d')


num_red_traj = 2

## Plot some trajectories red and thick
for sol in sols[:num_red_traj]:
    x_sol, y_sol, z_sol = sol[:, 0], sol[:, 1], sol[:, 2]
    ax.plot(x_sol, y_sol, z_sol, color = 'red', linewidth=0.8)

## Plot other trajectories blue
for sol in sols[num_red_traj:]:
    x_sol, y_sol, z_sol = sol[:, 0], sol[:, 1], sol[:, 2]
    ax.plot(x_sol, y_sol, z_sol, color = 'blue', linewidth=0.5)

## Make response graph
game.make_response_graph(fig, ax, plot_all = True)

## Plot specific points
#ax.plot( 0.5, 0.5, 0, 'ro' )
#ax.plot( 0.5, 0, 0.5, 'bo' )
ax.plot( 0.5, 0.5, 0.5, 'ko' )

## Plot zeros of payfield as CURVE given by sympy symbolic solution of the payfield system
ax.plot(*payfield_zeros_data_masked, 'orange')

## Plot zeros of payfield as SURFACE given normalized measures making the game harmonic
ax.plot_surface(Mu_Nash_x_surf, Mu_Nash_y_surf, Mu_Nash_z_surf)
#plt.savefig(f'./results/{name}.pdf', bbox_inches='tight')
```
