## Lambdify payfield

```python
payfield = sp.lambdify(vas, game.clean_payfield)
```

## Solve ODE, and plot trajectories

### Effective Dual averaging with entropic regularizer for [0,1]^N game

$$ \dot{x}_i = \frac{V_i(x)}{h''_i(x_i)} = x_i (1-x_i) V_i(x)$$

```python
def da(var, time):
        
    #x1, x2, x3 = var
    #V1, V2, V3 = payfield(x1, x2, x3)
        
    #F1 = sharp(x1) * V1
    #F2 = sharp(x2) * V2
    #F3 = sharp(x3) * V3
        
    return [ var[i] * (1 - var[i]) * payfield(*var)[i] for i in range(len(var)) ]
```

### Solve ODE

```python

from scipy.integrate import odeint

time = np.linspace(1, 1000, 50000)



# initial_points = [ (0.55, 0.45, 0.01), [0.45, 0.2, 0.55], [0.001, 0.001, 0.001]  ]  + [np.random.rand(3) for _ in range(2)]
initial_points = [np.random.rand(3) for _ in range(1)]
#initial_points = [ (0.51, 0.51, 0.01)]
#initial_points = [ (0.5, 0.1, 0.5)]
#initial_points = [ [0.5, 0.5, 0.5], [0.4, 0.2, 0.6] ]

sols = [ odeint(da, p, time) for p in initial_points ]
```

### Plot; response graph comes from nfg module; dynamics comes from here

```python
payfield_zeros
```

```python
name = 'perturbed-harmonic-zerosum'

fig = plt.figure(figsize=(18*1.3, 12*1.3))
ax = fig.add_subplot(1, 1, 1, projection='3d')


num_red_traj = 0

## Plot some trajectories red and thick
for sol in sols[:num_red_traj]:
    x_sol, y_sol, z_sol = sol[:, 0], sol[:, 1], sol[:, 2]
    ax.plot(x_sol, y_sol, z_sol, color = 'red', linewidth=0.8)

## Plot other trajectories blue
for sol in sols[num_red_traj:]:
    x_sol, y_sol, z_sol = sol[:, 0], sol[:, 1], sol[:, 2]
    ax.plot(x_sol, y_sol, z_sol, color = 'blue', linewidth=0.5)
    ax.plot(x_sol[0], y_sol[0], z_sol[0], 'bo')

## Make response graph
game.make_response_graph(fig, ax, plot_all = True)

## Plot specific points
#ax.plot( 0.5, 0.5, 0, 'ro' )
#ax.plot( 0.5, 0, 0.5, 'bo' )
#ax.plot( 0.5, 0.5, 0.5, 'ko' )

## Plot zeros of payfield as CURVE given by sympy symbolic solution of the payfield system
ax.plot(*payfield_zeros, 'ro')


#plt.savefig(f'./results/{name}.pdf', bbox_inches='tight')
```
