### Effective Dual averaging with entropic regularizer for [0,1]^N game

$$ \dot{x}_i = \frac{V_i(x)}{h''_i(x_i)} = x_i (1-x_i) V_i(x)$$

```python
def da(var, time):
        
    #x1, x2, x3 = var
    #V1, V2, V3 = payfield(x1, x2, x3)
        
    #F1 = sharp(x1) * V1
    #F2 = sharp(x2) * V2
    #F3 = sharp(x3) * V3
        
    return [ var[i] * (1 - var[i]) * payfield(*var)[i] for i in range(len(var)) ]
```

### Solve ODE

```python

from scipy.integrate import odeint

time = np.linspace(1, 1000, 100000)



#initial_points = [ (0.55, 0.45, 0.01), [0.45, 0.2, 0.55], [0.001, 0.001, 0.001]  ]  + [np.random.rand(3) for _ in range(20)]
initial_points = [np.random.rand(3) for _ in range(10)]
#initial_points = [ (0.51, 0.51, 0.01)]
#initial_points = [ (0.5, 0.1, 0.5)]

sols = [ odeint(da, p, time) for p in initial_points ]
```

### Plot; response graph comes from nfg module; dynamics comes from here

```python
list(game.pay_dict.keys())
```

```python
import matplotlib.pyplot as plt
from datetime import datetime
import aspera
SAVE = 1

fig = plt.figure(figsize=(9*1.5, 6*1.5))
ax = fig.add_subplot(1, 1, 1, projection='3d')


# number to plot red
plot_red = 0
for sol in sols[:plot_red]:
    x_sol, y_sol, z_sol = sol[:, 0], sol[:, 1], sol[:, 2]
    ax.plot(x_sol, y_sol, z_sol, color = 'red', linewidth=0.8)
    
for sol in sols[plot_red:]:
    x_sol, y_sol, z_sol = sol[:, 0], sol[:, 1], sol[:, 2]
    ax.plot(x_sol, y_sol, z_sol, color = 'blue', linewidth=0.5)
game.make_response_graph(fig, ax, plot_all = True)
#ax.plot( 0.5, 0.5, 0, 'ro' )
#ax.plot( 0.5, 0, 0.5, 'bo' )
#ax.plot( 0.5, 0.5, 0.5, 'ko' )
plt.tight_layout()


if SAVE:
    root = './Results/'
    now = datetime.now()
    timestamp = now.strftime("%Y-%m-%d_%H-%M-%S")
    day = now.strftime("%Y-%m-%d-")

    game_directory = f'{root}/{day + game.name}'
    current_directory = f'{game_directory}/{timestamp}'

    aspera.utils.make_folder(current_directory)
    aspera.utils.write_to_txt(f'{current_directory}/payoff.txt', f'{u_flat}\n' )
    aspera.utils.write_to_txt(f'{current_directory}/payoff.txt', game.pay_dict )
    plt.savefig(f'{current_directory}/{game.name}.pdf', bbox_inches='tight')#, pad_inches = 0)
```
