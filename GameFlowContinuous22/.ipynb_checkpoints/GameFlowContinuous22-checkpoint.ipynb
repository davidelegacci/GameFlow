## Regularizer

```python
#---------------------------------------------------------------------
# Define (symbolic) kernel functions
#---------------------------------------------------------------------

# 0 = sha
# 1 = log barrier
# 2 = q-replicator
# 3 = q-replicator

# q-replicator parameters
Q = [0.5] #, 2.5]

# Kiwiel kernel parameter; very similar to q-replicator
kiw = 0.5

# Teboulle kernel --> very similar to q-replicator

# sympy expressions
simkernels_nolog = [
    x * sp.log(x),
    (x**2 - x)/4,
    x * sp.log(x) - x,
    - (x**kiw) / kiw ] + [(x - x**q) / ( 1-q ) for q in Q ]

# kernel labels
kerlabels_nolog = ["Sha", "Eu",  "Bolt", "Kiw"] + [f"q = {q} Sha" for q in Q]
#---------------------------------------------------------------------

simkernels = simkernels_nolog # + [  - sp.log(x) ]
kerlabels = kerlabels_nolog # + [ "log" ]
```

```python
for i, theta in enumerate(simkernels):
    print(kerlabels[i], ":", theta)
```

```python
# Init class
A = legendre.LinFTRL(simkernels = simkernels, kerlabels = kerlabels, radius = 1)
```

## Games and dynamics

### Mixed extension of NFG

```python
class NFG():
    """docstring for ClassName"""
    def __init__(self, name, payoff_1, payoff_2):
        super(NFG, self).__init__()
        self.payoff_1 = payoff_1
        self.payoff_2 = payoff_2
        self.name = name


    def V(self, x1, x2):

        a1, b1, c1, d1 = self.payoff_1
        a2, b2, c2, d2 = self.payoff_2
    
        alpha1 = a1 + d1 - b1 - c1
        alpha2 = a2 + d2 - b2 - c2
        
        beta1 = b1 - d1
        gamma2 = c2 - d2
        
        return [alpha1 * x2 + beta1, alpha2 * x1 + gamma2] 
```

```python
# matching pennies
MP = NFG( "matching pennies", [1, -1, -1, 1], [-1, 1, 1, -1] )
```

```python
# PD
PD = NFG( "prisoner dilemma", [2, 0, 3, 1], [2, 3, 0, 1] )
```

## Continuous games

### Cournot competition

```python
cost = 1/4
cournot_eq = (1-cost)/3

def V_cournot(x1, x2):
    return [1 - cost - 2 * x1 - x2, 1 - cost - 2 * x2 - x1]
```

## Initial points

```python
time = np.linspace(0, 100, 100000)

epsilon = 1e-3

init_generator = np.linspace(epsilon, 1-epsilon, 6)

initial_points = [ p for p in product(init_generator, init_generator) ]
```

```python
def ftrl(var, time, sharp, payfield):
        
    x1, x2 = var
    V1, V2 = payfield(x1, x2)
        
    F1 = sharp(x1) * V1
    F2 = sharp(x2) * V2
        
    return [F1, F2]
```

## Experiments

```python
def make_plots(sharp_index, payfield, name, plot_point = False, point = 0):
    
    sols = [ odeint(ftrl, p, time, args=(A.sharps[sharp_index], payfield)) for p in initial_points ]
    
    [plt.plot(sol[:, 0], sol[:, 1], 'k', linewidth=0.5) for sol in sols]
    
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.title(f"{name} - {A.kerlabels[sharp_index]}")

    if plot_point:
        plt.plot(*point, "ko")
    plt.show()
    #return sols
        
    
```

```python
for i, h in enumerate(A.regularizers):
    make_plots(i, MP.V, "matching pennies")
```

```python
for i, h in enumerate(A.regularizers):
    make_plots(i, PD.V, "prisoners dilemma")
```

```python
for i, h in enumerate(A.regularizers):
    make_plots(i, V_cournot, "Cournot competition", plot_point = True, point = (cournot_eq, cournot_eq))
```
